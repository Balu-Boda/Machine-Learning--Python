```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# load the file in pandas DataFrame

```python
df =pd.read_csv("area.csv")
df
```

# once the DataFrame is ready plot a scatter plot

```python
# use this inorder to draw the plots (%matplotlib inline )
%matplotlib inline 
plt.scatter(df.area,df.price)
```

# make the scatter plot to be fancy

```python
plt.xlabel("area(sqft)")
plt.ylabel("price(USD)")
plt.scatter(df.area,df.price,color="green",marker="+")
```

# we can use liner regression now

```python
# first create LinearRegression Object
from sklearn import linear_model
reg = linear_model.LinearRegression()
# next fit the data --> means Training the LinearRegression model using the available data points
# first arguement should be 2D array
reg.fit(df[["area"]],df.price)
```

# now the model is ready to predict

```python
reg.predict([[4200]])
```

#

```python
reg.coef_
```

```python
reg.intercept_
```

# Area is availabe in the sheet and we need to predict price
 ** generate another csv file for further reference **

```python
d=pd.read_csv("area.csv")
d
```

# now predict the price

```python
reg.predict(d)
```

# we can send the results to CSV file

```python
p=reg.predict(d)
# create a new column and assign values 
d["price"]=p
d

```

# send/save the results to CSV file

```python
# send the results 
# to_csv("file name") and the file will be placed in the respective location
d.to_csv("new_prices.csv")
```
